```python
import numpy as np
import pandas as pd
from openbb import obb
obb.user.preferences.output_type = "dataframe"
```

```python
df = obb.equity.price.historical(
    "AAPL",
    start_date="2020-07-01",
    end_date="2023-07-06",
    provider="yfinance",
)
```

```python
calendar_dates = pd.date_range(
    start=df.index.min(),
    end=df.index.max(),
    freq="D"
)
```

```python
calendar_prices = df.reindex(calendar_dates)
```

```python
calendar_prices 
```

```python
# Replace nan values with the next valid observation
```

```python
calendar_prices.bfill()
```

```python
# ffill to propagate the last valid observation forward
calendar_prices.ffill()
```
