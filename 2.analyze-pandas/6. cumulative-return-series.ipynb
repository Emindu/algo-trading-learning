```python
import numpy as np
import pandas as pd
from openbb import obb
obb.user.preferences.output_type = "dataframe"
```

```python
df = obb.equity.price.historical(
    "AAPL",
    start_date="2020-01-01",
    provider="yfinance"
)
```

```python
close = df["close"]
```

```python
returns = close.pct_change()
```

```python
returns[np.isnan(returns)] = 0
```

```python
returns += 1
```

```python
cumulative_returns = returns.cumprod() - 1
```

```python
cumulative_returns.plot()
```

```python
## The cumulative sum of compound returns
```

```python
log_returns = np.log(close / close.shift())
```

```python
cumulative_log_returns = log_returns.cumsum()
```

```python
cumulative_log_returns.plot()
```

```python
	'''
    •	isnan Check: Identifies NaN values in the returns Series.
	•	Replacing NaNs: returns[np.isnan(returns)] = 0 replaces NaN values with 0.
	•	Incrementing Returns: returns += 1 adds 1 to each value in the returns Series to prepare for cumulative return calculation.
	•	Cumulative Product (cumprod): Calculates the cumulative product of the Series, multiplying values together sequentially.
	•	Example of cumprod: For a Series [1, 2, 3, 4], the cumulative product is [1, 2, 6, 24].
	•	Compute Compound Returns: Subtract 1 from the cumulative product to get compound returns.
	•	Cumulative Sum (cumsum): Computes the cumulative sum, adding values together sequentially instead of multiplying them.
    '''
```
