```python
import numpy as np
import pandas as pd
from openbb import obb
obb.user.preferences.output_type = "dataframe"
```

```python
df = obb.equity.price.historical(
    "AAPL",
    interval="1m",
    provider="yfinance"
)
```

```python
resampled = df.resample(rule="h")["close"]
```

```python
resampled.mean()
```

```python
resampled.ohlc()
```
